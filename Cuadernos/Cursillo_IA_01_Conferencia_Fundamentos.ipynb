```python
# Definir un tensor como variable con gradiente
x = torch.tensor(3.0, requires_grad=True)
z = torch.tensor(2.0, requires_grad=True)

# Realizar una operación
y = x * x +  z*z*z

# Calcular el gradiente (dy/dx)
y.backward()
dy_dx = x.grad
dy_dz = z.grad
print("Gradiente: dy_dx=", dy_dx)
print("Gradiente: dy_dz=", dy_dz)
```

## <span style="color:#4CC9F0">Entrenamiento de un perceptrón</span>

El problema de entranamiento de este perceptron es un problema de regresión.

Supongamos que tenemos la matriz de datos features)


$$
X = \begin{pmatrix} x_{11}& x_{12} & x_{13}\\
x_{21}& x_{22} & x_{23}\\
\ldots& \ldots & \ldots\\
x_{N1}& x_{N2} & x_{N3}\\
\end{pmatrix}
$$

y valores observados (targets) 

$$
Y = \begin{pmatrix} y_1\\
y_2\\
\ldots\\
y_N\\
\end{pmatrix}
$$

Por facilidad escibiremos $\boldsymbol{\beta} =(W,v,b,c)^t$. 

El perceptrón de arriba se escribirá $h_{\beta}(\cdot)$. 

Para vector cualquiera $x$, escribiremos $\hat{y}= h_{\beta}(x)= f(v'(g(Wx+b)=c)$

**Entrenar el perceptrón** de arriba para los datos $(X,Y)$ consiste en encontrar $\hat{\boldsymbol{\beta}}$ que minimiza la función de pérdida

$$
\mathfrak{L}(\boldsymbol{\beta}|X,Y) = \tfrac{1}{N}\sum_i (y_i - \hat{y}_i)^2 = \tfrac{1}{N}\sum_i (y_i - h_{\boldsymbol{\beta}}(x_i))^2 
$$

## <span style="color:#4CC9F0">Backpropagation</span>

Backpropagation, desarrollado en su forma actual en la década de 1980, es un algoritmo fundamental para el entrenamiento de redes neuronales. Aunque los conceptos básicos de backpropagation fueron propuestos por primera vez por Paul Werbos en su tesis doctoral en 1974, fue popularizado por el trabajo de Geoffrey Hinton, David Rumelhart y Ronald Williams en 1986. Su artículo "Learning representations by back-propagating errors" demostró cómo el algoritmo podía ser utilizado eficientemente para entrenar redes neuronales profundas, revolucionando el campo del aprendizaje automático y estableciendo las bases para el desarrollo de la inteligencia artificial moderna. 

Escencialmente, backpropagation es un algoritmo eficiente para calcular el gradiente de la función de pérdida. A continuación puede verse el algoritmo de optimización completo usando backpropagation.

### <span style="color:#4CC9F0">Algoritmo de Backpropagation para un Perceptrón con una Capa Oculta</span>

#### <span style="color:#4CC9F0">Notación</span>

- $ x $: Entrada.
- $ y $: Salida real.
- $ \hat{y} $: Salida predicha.
- $ W $: Pesos de la capa de entrada a la capa oculta.
- $ V $: Pesos de la capa oculta a la capa de salida.
- $ b $: Bias de la capa de entrada a la capa oculta.
- $ c $: Bias de la capa oculta a la capa de salida.
- $ \sigma $: Función de activación sigmoid, $\sigma(z) = \frac{1}{1 + e^{-z}}$.

#### <span style="color:#4CC9F0">Forward Pass (Propagación hacia adelante)</span> 

1. **Entrada a Capa Oculta**:
$$
   z^{(1)} = W \cdot x + b
  $$
$$
   a^{(1)} = \sigma(z^{(1)})
  $$

2. **Capa Oculta a Capa de Salida**:
$$
   z^{(2)} = V \cdot a^{(1)} + c
  $$
$$
   \hat{y} = \sigma(z^{(2)})
  $$

#### <span style="color:#4CC9F0">Cálculo del Error</span>

Se usa una función de pérdida, como el error cuadrático medio:
$$
E = \frac{1}{2} (y - \hat{y})^2
$$

#### <span style="color:#4CC9F0">Backward Pass (Propagación hacia atrás)</span> 

1. **Gradiente en la Capa de Salida**:
$$
   \delta^{(2)} = (\hat{y} - y) \cdot \sigma'(z^{(2)})
  $$
   Donde $\sigma'(z) = \sigma(z) \cdot (1 - \sigma(z))$.

2. **Gradiente en la Capa Oculta**:
$$
   \delta^{(1)} = (V^T \cdot \delta^{(2)}) \cdot \sigma'(z^{(1)})
  $$

#### <span style="color:#4CC9F0">Actualización de Pesos</span>

1. **Actualizar Pesos y Bias de la Capa Oculta a la Capa de Salida**:
$$
   V = V - \eta \cdot \delta^{(2)} \cdot (a^{(1)})^T
  $$
$$
   c = c - \eta \cdot \delta^{(2)}
  $$

2. **Actualizar Pesos y Bias de la Capa de Entrada a la Capa Oculta**:
$$
   W = W - \eta \cdot \delta^{(1)} \cdot x^T
  $$
$$
   b = b - \eta \cdot \delta^{(1)}
  $$

Donde $\eta$ es la tasa de aprendizaje.

## <span style="color:#4CC9F0">Diferenciación Automática y Backpropagation</span>

+ **Backpropagation** es un algoritmo específico para calcular los gradientes de la función de pérdida con respecto a los pesos de una red neuronal. Es esencialmente una aplicación de la regla de la cadena para derivadas parciales, aplicada de manera eficiente a través de las capas de la red neuronal.

+ **Diferenciación Automática** es una técnica general para evaluar derivadas de funciones especificadas por programas informáticos. Existen dos modos principales de diferenciación automática:

+ **Relación entre Ambos**. En la práctica, las librerías modernas de machine learning, como TensorFlow y PyTorch, utilizan autodiff para implementar backpropagation. La autodiff permite calcular los gradientes necesarios de manera automática y eficiente sin que el usuario tenga que derivar manualmente las funciones.

### <span style="color:#4CC9F0">Gradiente Descendiente Estocástico</span>

El descenso de gradiente estocástico (SGD), realiza una actualización de parámetros para cada ejemplo de entrenamiento $x_{train}^{(i)} $ y etiqueta $ y_{train}^ {(i)} $, **seleccionados al azar en cada época**.


$$
\boldsymbol{\beta}_{k+1} =  \boldsymbol{\beta}_k - \eta_k \nabla_{\boldsymbol{\beta}} \mathfrak{L}({x}_{train}^{(i)},{y}_{train}^{(i)},\boldsymbol{\beta}_k),
$$


En el artículo original de [Robbins and Monro (1951)](https://projecteuclid.org/download/pdf_1/euclid.aoms/1177729586) $\eta_k$ cambia en cada iteración y los autores probaron que si se asume que  $\{\eta_k\}$ es una sucesión tal que $\sum_k \eta_k = \infty$, and $\sum_k \eta_k^2 < \infty$, la sucesión  $\boldsymbol{\beta}_k$ coverge a la solución del problema con probabilidad 1. Por ejemplo, se puede escoger $\eta_k = 1/k$. Robbins y Monro demostraron el resultado bajo condiciones muy generales.

### <span style="color:#4CC9F0">Gradiente Descendiente Estocástico por minilotes</span>

El descenso de gradiente por mini-lotes realiza una actualización para cada mini-lote de $n$ ejemplos de entrenamiento:


$$
\boldsymbol{\beta}_{k+1} =  \boldsymbol{\beta}_k - \eta_k \nabla_{\boldsymbol{\beta}} \mathfrak{L}(\mathbf{x}_{train}^{(i:i+n)},\mathbf{y}_{train}^{(i:i+n)},\boldsymbol{\beta}_k),
$$



Desde este punto de la lección, asumiremos que **tomamos mini-lotes**, por lo que omitimos súper-índices en los datos $(\mathbf{x}_{train}^{(i:i+n)},\mathbf{y}_{train}^{(i:i+n)})$ en todas las expresiones.

Matemáticamente, el modelo puede escribirse en la forma

$$
\begin{align}

$$



De forma muy general podemos escribir el perceptron multicapa de la forma

$$
\boldsymbol{y} = g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x})
$$

# <span style="color:#4361EE">Regresión con una Red Profunda</span> 

Par esta sección supondremos  el perceptrón multicapa $g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}): \mathcal{R}^n \to \mathcal{R}^m$. 

La matriz de features observados es

$$
\boldsymbol{X} = \begin{pmatrix} x_{11}& x_{12} & \ldots&x_{1n}\\
x_{21}& x_{22} & \ldots&x_{2n}\\
\ldots& \ldots & \ldots\\
x_{N1}& x_{N2} & \ldots&x_{Nn}\\
\end{pmatrix}.
$$

La matriz de targets observados es

$$
\boldsymbol{Y} = \begin{pmatrix} y_{11}& y_{12} & \ldots&y_{1m}\\
y_{21}& x_{22} & \ldots&y_{2m}\\
\ldots& \ldots & \ldots\\
y_{N1}& y_{N2} & \ldots&y_{Nm}\\
\end{pmatrix}.
$$

**Entrenar el perceptrón** para los datos $(\boldsymbol{X},\boldsymbol{Y})$ consiste en encontrar $(\widehat{\mathbf{W}}, \hat{\mathbf{b}})$ que minimiza la función de pérdida

$$
\mathfrak{L}((\mathbf{W},\mathbf{b})|\boldsymbol{X},\boldsymbol{Y}) = \tfrac{1}{N}\sum_{i=1}^N ||(y_{i} - \hat{y}_{i}||^2 = \tfrac{1}{N}\sum_i ||y_i - g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}_i)||^2 
$$

## <span style="color:#4CC9F0">Regularización</span>

La función $g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x})$ por lo general tiene múltiples mínimos locales. Nunca es está seguro de si se ha encontrado un mínimo global. En lugar de eso, se busca un solución que sea aceptable para los propósitos del problema. Un efecto no deseado es el sobreajuste, que ocurre cuando $\mathbf{W}$, visto como un vector, tiende a tener una norma muy grande. Para evitar el problema se introducen restricciones conocidas como regularizaciones. La regularización $L_2$ consiste en agregar el término $\lambda ||\mathbf{W}||^2$  a la función de pérdida. $\lambda$ es un hiperparámetro entre  0 y 1. De esta forma la función de pérdida resulta siendo

$$
\mathfrak{L}((\mathbf{W},\mathbf{b})|\boldsymbol{X},\boldsymbol{Y}) = \tfrac{1}{N}\sum_i ||y_i - g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}_i)||^2 + \lambda ||\mathbf{W}||^2.
$$

# <span style="color:#4361EE">Clasificación con una Red Profunda</span> 

Par esta sección supondremos  el perceptrón multicapa 
$g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}): \mathcal{R}^n \to \mathcal{R}^J$. De momento no hay nada especial con el perceptron. Se tiene un problema de clasificación si los targets $y \in \{1,\ldots,J\}$. En este caso se tiene un problema de clasificación con $J$ clases. 

Por facilidad consideramos primero el problema de clasificación en dos clases.


## <span style="color:#4CC9F0">Perceptrón de clasificación en dos clases</span>

En este caso el perceptron es un función de valor real $g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}): \mathcal{R}^n \to \mathcal{R}$. Es usual, pero no extrictamente necesario que la función de  la última capa sea el sigmoide, con lo cual $g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x})$ se puede interpretar como la probabilidad de que $\boldsymbol{x}$ sea de la clase 1. Supondremos entonces que la activación de la  última capa es el sigmoide.

El proceso de entrenamiento del perceptrón consiste en estimar los parámetros $(\mathbf{w},\mathbf{b})$ del modelo a partir de los datos de entrenamiento recibidos.

La función de probabilidad definida por el modelo de Bernoulli es dada por

$$
P(Y=y| \mathbf{w},\mathbf{b}, \mathbf{x}) = [g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})]^y \times [1-g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})]^{1-y}, \hspace{3mm} y\in \{0,1\}.
$$


Al disponer de un dato de entrenamiento nuestra función ahora tiene conocidos los valores $(\mathbf{x},y)$. Vamos a definir la función de verosimilitud con la misma expresión, pero ahora consideramos como desconocidos a los parámetros $\mathbf{w}$ y $\mathbf{b}$. Esta nueva función se denomina veromilitud para la observación $(\mathbf{x},y)$ y se denota

$$
l(\mathbf{W},\mathbf{b}|\mathbf{x},y) =  [g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})]^y \times [1-g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})]^{1-y}
$$

Si tomamos logaritmo y anteponemos el signo menos obtenemos la función de pérdida

$$
\mathfrak{L}(\mathbf{W},\mathbf{b}|\mathbf{x},y) = - (y\log[g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})] + (1-y)[\log[1-g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})])
$$

Hemos encontrado la entropía cruzada entre dos distribuciones:

+ La distribución de referencia dada por $[y, 1-y]$
+ La distribución aproximante dada por $[g_{(\mathbf{W},\mathbf{b})}(\mathbf{x}), 1-g_{(\mathbf{W},\mathbf{b})}(\mathbf{x})]$.

Para el conjunto de datos de entrenamiento $(\mathbf{x}_i, y_i), \hspace{3mm} i=1,\ldots,N$, la entropía cruzada *promedio* es dada por

$$
\mathfrak{L}(\mathbf{W},b|\mathbf{X},\mathbf{y}) = -\frac{1}{N} \sum_{i=1}^N y_i\log[g_{(\mathbf{W},b)}(\mathbf{x}_i)] + (1-y_i)[\log[1-g_{(\mathbf{W},b)}(\mathbf{x}_i)]
$$

Esta es la función de pérdida, la entropía cruzada, que es necesario minimizar con respecto a $\mathbf{W}$ y $\mathbf{b}$.



## <span style="color:#4CC9F0">Modelo de clasificación en $J$ clases</span>


Para este caso el perceptron es dado por $g_{(\mathbf{W},\mathbf{b})}(\boldsymbol{x}): \mathcal{R}^n \to \mathcal{R}^J$. El vector de salida $Y$ es llamado vector de prepobabilidades. De nuevo nada especial con el perceptron. Para interperetar cada componente como una probabilidad calculamos la función softmax de $Y$, con lo cual, el vector de probabilidades de salida es

$$
P = softmax(Y),
$$
en donde $P=(p_1, , \ldots, p_J)$ y cada $p_j$ es dado por

$$
p_j = \frac{e^{y_j}}{\sum_k e^{y_k}}.
$$

Se tiene entonces que $P$ representa la distribución de probabilidades para el respectivo vector $\boldsymbol{x}$.



Ahora usaremos el truco `one-hot`, para emparejar el target con la salida del perceptron. Definimos el tensor $Z = (Z_1, \ldots, Z_j, \ldots Z_J)$, en donde

$$
Z_j = \begin{cases} &1, \quad \text{si } Y=j\\
& 0, \quad \text{en otro caso} 
\end{cases}
$$




Es decir, $Z$ es el vector que representa la codificación `one-hot` de la etiqueta $Y$. Observe que $Z$ solamente tiene 1 en la posición $j$, si la categoría de la etiqueta es $j$, y cero en el resto de posiciones.

Finalmente, para este caso la función de pérdida es la entropía cruzada, definida para el caso general de $N$ muestras en un lote de datos como


$$
\mathfrak{L} = -\frac{1}{N} \sum_{i=1}^N\left[ \sum_{j=1}^J  Z_{ij}log (\hat{p}_{ij}) \right],
$$

en donde $\hat{Y}_i= (\hat{Y}_{i1}, \ldots, \hat{Y}_{iJ})$ es la salida del la red para la entrada $X_i$, $\hat{p}_i = \text{softmax}(\hat{Y}_i)$, y $Z_i$ la codificación *one-hot* de la etiqueta $Y_i$.

#  <span style="color:blue">Funciones de Pérdida</span>

## <span style="color:#4CC9F0">Funciones de pérdida para regresión</span>

En los problemas de regresión se supone que tanto la salida como la etiqueta con tensores de tipo *float*. En este caso algunas de las funciones de pérdida más utilizadas son:

+ Error cuadrático medio univariante: $$\text{ECM} = \frac{1}{N} \sum_{i=1}^N (y_i-\hat{y}_i)^2$$
+ Error cuadrático medio multivariante: $$
\text{ECM} = \frac{1}{N} \sum_{i=1}^N \sum_{j=1}^J (y_{ij}-\hat{y}_{ij})^2
$$
+ Error absoluto medio univariate: $$
\text{EAM} = \frac{1}{N} \sum_{i=1}^N |y_i-\hat{y}_i|
$$
+ Error absoluto medio multivariante: $$
\text{EAM} = \frac{1}{N} \sum_{i=1}^N \sum_{j=1}^J|y_{ij}-\hat{y}_{ij}|
$$
+ Función de perdida de Hubber unvariante: $$
\begin{align}
\text{Hubber loss} &= \tfrac{1}{n}\sum_{i=1}^n (Y_i- \hat{Y}_i)^2 \hspace{2cm} |\hat{Y}_i-Y_i|\le \delta\\
&=  \tfrac{1}{n}\sum_{i=1}^n |Y_i- \hat{Y}_i)| \hspace{2cm} |\hat{Y}_i-Y_i|> \delta
\end{align}
$$
+ Función de perdida de Hubber multivariante: $$
\begin{align}
\text{Hubber loss} &= \tfrac{1}{N}\sum_{i=1}^N \sum_{j=1}^J(Y_{ij}- \hat{Y}_{ij})^2 \hspace{2cm} \sum_{j=1}^J|\hat{Y}_{ij}-Y_{ij}|\le \delta\\
&=  \tfrac{1}{N}\sum_{i=1}^N \sum_{j=1}^J |Y_{ij}- \hat{Y}_{ij})| \hspace{2cm} \sum_{j=1}^J|\hat{Y}_{ij}-Y_{ij}|> \delta
\end{align}
$$

## <span style="color:#4CC9F0">Funciones de pérdida para clasificación</span>

En los problemas de clasificación, las etiquetas son discretas. Cada ejemplo de entrenamiento tiene una categoría asociada.

En la práctica $p_j$ se interpreta como la probabilidad que la entrada $X$ pertenezca a la clase $j$.

+ Entropía cruzada binaria: $$
\mathfrak{L} = -\left[(\log (\hat{p}) Y) + \log (1-\hat{p})(1-Y) \right]$$.

+ Entropía cruzada, $J$ categorías:
$$
\mathfrak{L} = -\frac{1}{N} \sum_{i=1}^N\left[ \sum_{j=1}^J  Z_{ij}log (\hat{p}_{ij}) \right],
$$


#  <span style="color:blue">Métricas</span>

Las métricas son estadísticas quew se usan usualmente como criterio de evaluación de la calidd del entrenamiento. ALGUNAS DE LA smáas usadas son:

- **accuracy_score**: En la clasificación con múltiples etiquetas, esta función calcula la precisión del subconjunto: el conjunto de etiquetas predichas para una muestra que coincide exactamente con el conjunto de etiquetas correspondiente en y_true.
- **precision_score**: es la razón $\frac{tp }{tp + fp}$ en donde $tp$ es el número de positivos verdadero y $fp$ el número de falsos positivos. El mejor valor es 1 y el peor valor es 0.
- **recall_score**:  es la relación $\frac{tp }{tp + fn}$ donde $tp$ es el número de verdaderos positivos y $fn$ el número de falsos negativos. El recuerdo es intuitivamente la capacidad del clasificador para encontrar todas las muestras positivas. El mejor valor es 1 y el peor valor es 0.
- **ecm**: $\text{ECM} = \frac{1}{N} \sum_{i=1}^N (y_i-\hat{y}_i)^2$

#  <span style="color:blue">Notas Finales</span>

1. La mayor parte de los modelos estadísticos clasicos son casos particulares del un perceptron sin capa oculta de una neurona.
2. Con probabilidad 1 y bajo condiciones muy generales, cualquier función $f: \mathcal{R}^n \to \mathcal{R}^m$ puede ser aproximada por un perceptron multicapa, con un error máximo $\epsilon$ predefinido. Este resultado es el **teorema de aproximación universal con perceptrones**.
3. En la práctica, el conjunto de datos se divide en tres subconjuntos: entrenamieneto, validación interna y prueba. Los dos primeros se usan durante el entrenamiento para evaluar la calidad y el poder de generalizacion del perceptron y el último para evaluar externamente el modelo. Usualmente se utiliza la regla 80%, 10%, 10%, para dividir ewl conunto de datos.
